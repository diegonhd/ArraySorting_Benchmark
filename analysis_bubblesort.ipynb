```python
import numpy as np
import matplotlib.pyplot as plt
```

## Carregando os arquivos

```python
# Carregando um array de cada tamanho
arr_1k = np.load('archive/arr1_1k.npy')
arr_5k = np.load('archive/arr1_5k.npy')
arr_10k = np.load('archive/arr1_10k.npy')
arr_50k = np.load('archive/arr1_50k.npy')
```

```python
# Carregando os piores tempos de ordenação
time_wc_1k = np.load('archive/time_wc_1k.npy')
time_wc_5k = np.load('archive/time_wc_5k.npy')
time_wc_10k = np.load('archive/time_wc_10k.npy')
time_wc_50k = np.load('archive/time_wc_50k.npy')
```

```python
# Carregando as listas dos tempos de cada array aleatório
list_arr_1k = np.load('archive/list_arr_1k.npy')
list_arr_5k = np.load('archive/list_arr_5k.npy')
list_arr_10k = np.load('archive/list_arr_10k.npy')
list_arr_50k = np.load('archive/list_arr_50k.npy')
```

```python
# Carregando as médias dos tempos
mean_time_1k = np.load('archive/mean_arr_1k.npy')
mean_time_5k = np.load('archive/mean_arr_5k.npy')
mean_time_10k = np.load('archive/mean_arr_10k.npy')
mean_time_50K = np.load('archive/mean_arr_50k.npy')
```

## Média x Pior Caso Em Cada 'N' (Tamanho Da Lista)

### N = 1k

```python
y = np.hstack(([mean_time_1k, time_wc_1k], list_arr_1k))
cores = ['blue', 'red', 'green', 'orange', 'purple']
categorias = np.array(['Média', 'Pior Caso', 'Aleatório 1', 'Aleatório 2', 'Aleatório 3'])
fig, ax = plt.subplots(figsize = (5, 2.7), layout = 'constrained')
ax.bar(categorias, y, color = cores)
ax.set_ylabel('Tempo(s)')
ax.set_title("Comparação interna entre os arrays de 1k")
```

### N = 5k

```python
y = np.hstack(([mean_time_5k, time_wc_5k], list_arr_5k))
cores = ['blue', 'red', 'green', 'orange', 'purple']
categorias = np.array(['Média', 'Pior Caso', 'Aleatório 1', 'Aleatório 2', 'Aleatório 3'])
fig, ax = plt.subplots(figsize = (5, 2.7), layout = 'constrained')
ax.bar(categorias, y, color = cores)
ax.set_ylabel('Tempo(s)')
ax.set_title("Comparação interna entre os arrays de 5k")
```

### N = 10k

```python
y = np.hstack(([mean_time_10k, time_wc_10k], list_arr_10k))
cores = ['blue', 'red', 'green', 'orange', 'purple']
categorias = np.array(['Média', 'Pior Caso', 'Aleatório 1', 'Aleatório 2', 'Aleatório 3'])
fig, ax = plt.subplots(figsize = (5, 2.7), layout = 'constrained')
ax.bar(categorias, y, color = cores)
ax.set_ylabel('Tempo(s)')
ax.set_title("Comparação interna entre os arrays de 10k")
ax.set_ylim(0, 30)
```

```python
y = np.hstack(([mean_time_50K, time_wc_50k], list_arr_50k))
cores = ['blue', 'red', 'green', 'orange', 'purple']
categorias = np.array(['Média', 'Pior Caso', 'Aleatório 1', 'Aleatório 2', 'Aleatório 3'])
fig, ax = plt.subplots(figsize = (5, 2.7), layout = 'constrained')
ax.bar(categorias, y, color = cores)
ax.set_ylabel('Tempo(s)')
ax.set_title("Comparação interna entre os arrays de 50k")
ax.set_ylim(0, 800)
```

### Gráfico de Linhas 

```python
list_len = np.array([1000, 5000, 10000, 50000])
mean_times = np.array([mean_time_1k, mean_time_5k, mean_time_10k, mean_time_50K])
worst_case_times = np.array([time_wc_1k, time_wc_5k, time_wc_10k, time_wc_50k])
rand_arr1 = np.array([list_arr_1k[0], list_arr_5k[0], list_arr_10k[0], list_arr_50k[0]])
rand_arr2 = np.array([list_arr_1k[1], list_arr_5k[1], list_arr_10k[1], list_arr_50k[1]])
rand_arr3 = np.array([list_arr_1k[2], list_arr_5k[2], list_arr_10k[2], list_arr_50k[2]])
```

```python
fig, ax = plt.subplots(figsize = (5,2.7), layout = 'constrained')

ax.plot(list_len, mean_times, 
        label= 'Médias dos tempos')

ax.plot(list_len, worst_case_times,
        label= 'Piores tempos possíveis')

ax.plot(list_len, rand_arr1,
        label = 'Linha Aleatória 1')

ax.plot(list_len, rand_arr2,
        label = 'Linha Aleatória 2')

ax.plot(list_len, rand_arr3,
        label = 'Linha Aleatória 3')

ax.set_xlim(1000, 50000) 
ax.set_xlabel('Tamanho do array')
ax.set_ylabel('Tempos (s)')
ax.set_title("Tamanho x Tempos")
ax.legend()
ax.grid(True)
```
