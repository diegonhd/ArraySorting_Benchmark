## Definindo as funções usadas no código

```python
# Definindo a função bubble sort sem lib (obrigatório)
def bubble_sort(arr):
    lista = np.array(arr)
    n = len(lista)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
```

```python
def selection_sort(arr):
    lista = np.array(arr.copy())
    n  = len(lista)
    for j in range(n-1):
        min_index = j
        for i in range(j, n):
            if lista[i] < lista[min_index]:
                min_index = i
        lista[j], lista[min_index] = lista[min_index], lista[j]
```

```python
def insertion_sort(arr):
    lista = np.array(arr.copy())
    n = len(lista)
    for i in range(n):
        cur = lista[i]
        j = i - 1
        while j >= 0 and lista[j] > cur:
            lista[j+1], lista[j] = lista[j], cur
            j -= 1
    return lista
```

```python
def shell_sort(arr):
    lista = np.array(arr.copy())
    n = len(lista)
    gap = n // 2
    while gap >= 1:
        for i in range(n):
            cur = lista[i]
            j = i - gap
            while j >= 0 and lista[j] > cur:
                lista[j+gap], lista[j] = lista[j], cur
                j-= gap
        print(lista)
        gap //= 2
```

```python
def quick_sort(arr):
    lista = arr.copy()
    n = len(lista)
    if len(lista) <= 1:
        return np.array(lista)
    else:
        pivo = lista[n // 2]
        menores = [x for x in lista if x < pivo]
        iguais = [x for x in lista if x == pivo]
        maiores = [x for x in lista if x > pivo]
        return np.hstack([quick_sort(menores), iguais, quick_sort(maiores)]).astype(np.int32)
```

```python
arr = np.random.randint(1,101,20)
def merge_sort(arr, inicio=0, fim=None):
    lista = arr
    if fim is None:
        fim = len(lista)
    if (fim - inicio) > 1:
        meio = (inicio + fim) // 2
        merge_sort(arr, inicio, meio)
        merge_sort(arr, meio, fim)
        merge(lista, inicio, meio, fim)
    return lista

def merge(arr, inicio, meio, fim):
    lista = arr
    esquerda = lista[inicio:meio].copy()
    direita = lista[meio:fim].copy()
    i = 0
    j = 0 
    # i para esquerda j para direita
    for k in range(inicio, fim):
        if i >= len(esquerda):
            lista[k] = direita[j]
            j += 1
        elif j >= len(direita):
            lista[k] = esquerda[i]
            i += 1
        elif esquerda[i] <= direita[j]:
            lista[k] = esquerda[i]
            i += 1
        elif direita[j] < esquerda[i]: #podia ser else, mas quis ilustrar melhor a condição
            lista[k] = direita[j]
            j += 1

merge_sort(arr)
```

```python
# Definindo a função de retorno do tempo que o array levou a ser ordenado
def tempo_ordenaçao(lista):
    start_time = time.time()
    bubble_sort(lista)
    end_time = time.time()
    total_time = end_time - start_time
    return total_time
```

## Para o caso dos arrays de size = 1k

```python
# Gerando os 3 arrays aleatórios de cada caso (1k)
arr1_1k = np.random.randint(1,1001,1000)
arr2_1k = np.random.randint(1,1001,1000)
arr3_1k = np.random.randint(1,1001,1000)
arr_1k_sorted = np.random.randint(1,1001,1000)

# Fazendo um array do pior caso 
arr_wc_1k = np.sort(arr1_1k.copy())[::-1]
```

```python
# Pegando o tempo de ordenação dos 4 arrays
time_arr1_1k =  tempo_ordenaçao(arr1_1k)
print(f"Tempo do primeiro array de 1k: {time_arr1_1k} segundos")
time_arr2_1k = tempo_ordenaçao(arr2_1k)
print(f"Tempo do primeiro array de 1k: {time_arr2_1k} segundos")
time_arr3_1k = tempo_ordenaçao(arr3_1k)
print(f"Tempo do primeiro array de 1k: {time_arr3_1k} segundos")
time_wc_1k = tempo_ordenaçao(arr_wc_1k)
print(f"Tempo do primeiro array de 1k: {time_wc_1k} segundos")
```

```python
# Fazendo uma média dos 3 casos dos arrays aleatórios
list_arr_1k = np.array([time_arr1_1k, time_arr2_1k, time_arr3_1k])
mean_arr_1k = list_arr_1k.mean()
print(f"Tempo médio de ordenação dos arrays de 1k: {mean_arr_1k} segundos.")
```
